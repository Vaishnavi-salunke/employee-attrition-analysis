## Business Problem

Employee attrition is a major challenge for organizations.
This project aims to analyze employee data to understand
the key factors that contribute to employee attrition and
provide actionable insights to reduce turnover.

```python
df.groupby('Attrition')['MonthlyIncome'].mean()
```

“Employees who left the organization had a significantly lower average monthly income compared to retained employees, indicating compensation as a key driver of attrition.”

```python
df.groupby('Attrition')['Age'].mean()
```

### Age vs Attrition Insight

Employees who left the organization are younger on average (33.6 years) compared to employees who stayed (37.6 years). This indicates higher attrition among early-career professionals.


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='Age', data=df)
plt.title('Age vs Attrition')
plt.show()
```

```python
df.groupby('Attrition')['YearsInCurrentRole'].mean()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='YearsInCurrentRole', data=df)
plt.title('Years in Current Role vs Attrition')
plt.show()
```

## Key Business Insights

1. Employees who left the organization earn significantly lower monthly income compared to retained employees, highlighting compensation as a major attrition driver.

2. Younger employees show higher attrition rates, indicating early-career churn.

3. Employees with fewer years in their current role are more likely to leave, suggesting role mismatch or lack of growth opportunities.

4. Longer role tenure is associated with employee retention.


## Recommendations

1. Review compensation strategies for early-career employees to reduce attrition.

2. Improve onboarding and role clarity to reduce early dissatisfaction.

3. Introduce structured career growth plans and internal mobility programs.

4. Monitor employees with low role tenure as high-risk attrition candidates.
